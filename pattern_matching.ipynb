```python
def gerer_ticket(priorite):
    if priorite == "haute":
        return "Traitement immediat"
    elif priorite == "moyenne":
        return "Traiter dans les 24 heures"
    elif priorite == "basse":
        return "Traiter dans les 72 heures"
    else:
        return "priorite inconnue"
```

```python
def gerer_ticket_nouv(priorite):
    match priorite:
        case "haute":
            return "Traitement immediat"
        case "moyenne":
            return "Traiter dans les 24 heures"
        case "basse":
            return "Traiter dans les 72 heures"
        case _:
            return "priorite inconnue"

        
```

```python
print(gerer_ticket_nouv("haute"))
print(gerer_ticket_nouv("basse"))
print(gerer_ticket_nouv("moyenne"))
print(gerer_ticket_nouv("inconnue"))
```
