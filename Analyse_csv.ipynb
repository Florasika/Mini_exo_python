```python
#  Analyseur de Fichier CSV

import csv

# ================================
#  LIRE LE FICHIER CSV
# ================================

produits = []

with open(r"C:\Users\maath\Downloads\ventes.csv", newline='', encoding='utf-8') as fichier:
    lecteur = csv.DictReader(fichier)       # chaque ligne ‚Üí dictionnaire

    for ligne in lecteur:
        ligne['prix']     = float(ligne['prix'])
        ligne['quantite'] = int(ligne['quantite'])
        produits.append(ligne)

print(f" {len(produits)} produits charg√©s !\n")

# ================================
#  CALCULER LA MOYENNE DES PRIX
# ================================

prix_total = 0

for p in produits:
    prix_total += p['prix']

moyenne = prix_total / len(produits)
print(f" Prix moyen : {moyenne:.2f}‚Ç¨")
print(f" Prix min   : {min(p['prix'] for p in produits):.2f}‚Ç¨")
print(f" Prix max   : {max(p['prix'] for p in produits):.2f}‚Ç¨")

# set() supprime les doublons, sorted() trie alphab√©tiquement
categories = sorted(set(p['categorie'] for p in produits))

print("\n Cat√©gories disponibles :")
for i, cat in enumerate(categories, start=1):
    print(f"  {i} ‚Üí {cat}")

# ================================
# L'UTILISATEUR CHOISIT
# ================================

choix = input("\nEntrez le num√©ro de la cat√©gorie : ")

try:
    index            = int(choix) - 1          # -1 car les listes commencent √† 0
    categorie_voulue = categories[index]        # on r√©cup√®re la cat√©gorie choisie
except (ValueError, IndexError):
    print(" Num√©ro invalide !")
    exit()

# ================================
#  FILTRER ET AFFICHER
# ================================

print(f"\n Produits dans '{categorie_voulue}' :")
print("-" * 40)

for p in produits:
    if p['categorie'] == categorie_voulue:
        print(f"  ‚Ä¢ {p['produit']} ‚Üí {p['prix']}‚Ç¨")
```
