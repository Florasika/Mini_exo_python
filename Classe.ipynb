```python
class CompteBancaire:
    def __init__(self,numeroCompte, nom, solde):
        self.numeroCompte = numeroCompte
        self.nom = nom
        self.solde = solde
        
    def versement(self, argent):
        self.solde = self.solde + argent

    def retrait(self, argent):
        if(self.solde < argent):
            print("le solde est insufissant")
        else:
            self.solde = self.solde - argent

    def agios(self):
        self.solde = self.solde*95/100

    def afficher(self):
        print("Compte numéro :" ,self.numeroCompte)
        print("Nom :" ,self.nom)
        print("Solde :" ,self.solde)
        
```

```python
monCompte = CompteBancaire(111000253,"Alien",25000)
```

```python
monCompte.versement(5000)
```

```python
monCompte.retrait(1000)
```

```python
monCompte.afficher()
```
