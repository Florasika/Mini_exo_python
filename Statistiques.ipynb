```python
#  Statistiques Descriptives Manuelles

import math

# ================================
#  FONCTIONS STATISTIQUES
# ================================

def moyenne(donnees):
    """Somme de tous les Ã©lÃ©ments divisÃ©e par leur nombre"""
    return sum(donnees) / len(donnees)


def mediane(donnees):
    """Valeur centrale une fois les donnÃ©es triÃ©es"""
    tries = sorted(donnees)
    n     = len(tries)
    milieu = n // 2

    if n % 2 == 1:                          # nombre impair â†’ valeur du milieu
        return tries[milieu]
    else:                                   # nombre pair â†’ moyenne des 2 du milieu
        return (tries[milieu - 1] + tries[milieu]) / 2


def mode(donnees):
    """Valeur qui apparaÃ®t le plus souvent"""
    compteur = {}
    for valeur in donnees:
        compteur[valeur] = compteur.get(valeur, 0) + 1

    max_count = max(compteur.values())
    modes     = [k for k, v in compteur.items() if v == max_count]
    return sorted(modes)


def variance(donnees):
    """Moyenne des carrÃ©s des Ã©carts Ã  la moyenne"""
    moy    = moyenne(donnees)
    ecarts = [(x - moy) ** 2 for x in donnees]   # carrÃ© de chaque Ã©cart
    return sum(ecarts) / len(donnees)


def ecart_type(donnees):
    """Racine carrÃ©e de la variance"""
    return math.sqrt(variance(donnees))


def etendue(donnees):
    """DiffÃ©rence entre le max et le min"""
    return max(donnees) - min(donnees)


def quartiles(donnees):
    """Divise les donnÃ©es en 4 parties Ã©gales"""
    tries = sorted(donnees)
    n     = len(tries)

    q2 = mediane(tries)                         # mÃ©diane = Q2

    moitie_basse = tries[:n // 2]               # premiÃ¨re moitiÃ©
    moitie_haute = tries[n // 2 + n % 2:]       # deuxiÃ¨me moitiÃ©

    q1 = mediane(moitie_basse)
    q3 = mediane(moitie_haute)

    return q1, q2, q3


# ================================
# 2ï¸âƒ£ AFFICHAGE DU RAPPORT
# ================================

def afficher_rapport(donnees):
    """Calcule et affiche toutes les statistiques"""

    moy   = moyenne(donnees)
    med   = mediane(donnees)
    mod   = mode(donnees)
    var   = variance(donnees)
    et    = ecart_type(donnees)
    ext   = etendue(donnees)
    q1, q2, q3 = quartiles(donnees)

    print("\n" + "=" * 45)
    print("    STATISTIQUES DESCRIPTIVES")
    print("=" * 45)
    print(f"   DonnÃ©es    : {sorted(donnees)}")
    print(f"   Effectif   : {len(donnees)} valeurs")
    print("â”€" * 45)
    print(f"   Moyenne    : {moy:.2f}")
    print(f"   MÃ©diane    : {med:.2f}")
    print(f"   Mode(s)    : {mod}")
    print("â”€" * 45)
    print(f"   Minimum    : {min(donnees)}")
    print(f"   Maximum    : {max(donnees)}")
    print(f"   Ã‰tendue    : {ext}")
    print("â”€" * 45)
    print(f"   Variance   : {var:.2f}")
    print(f"   Ã‰cart-type : {et:.2f}")
    print("â”€" * 45)
    print(f"   Q1 (25%)   : {q1:.2f}")
    print(f"   Q2 (50%)   : {q2:.2f}  â† mÃ©diane")
    print(f"   Q3 (75%)   : {q3:.2f}")
    print(f"   IQR        : {q3 - q1:.2f}")
    print("=" * 45)


def afficher_histogramme(donnees, nb_barres=5):
    """Affiche un histogramme ASCII des donnÃ©es"""
    minimum = min(donnees)
    maximum = max(donnees)
    largeur = (maximum - minimum) / nb_barres

    print("\n   HISTOGRAMME")
    print("  " + "â”€" * 38)

    for i in range(nb_barres):
        borne_basse = minimum + i * largeur
        borne_haute = borne_basse + largeur
        count = sum(1 for x in donnees if borne_basse <= x < borne_haute)

        # derniÃ¨re barre : inclure le maximum
        if i == nb_barres - 1:
            count = sum(1 for x in donnees if borne_basse <= x <= borne_haute)

        barre = "" * count
        print(f"  [{borne_basse:5.1f} â€“ {borne_haute:5.1f}]  {barre:<15} ({count})")

    print("  " + "â”€" * 38)


# ================================
#  SAISIE UTILISATEUR
# ================================

print("=" * 45)
print(" STATISTIQUES DESCRIPTIVES")
print("=" * 45)
print("  Entrez vos nombres sÃ©parÃ©s par des espaces")

try:
    entree  = input("  â†’ ").strip()
    donnees = [float(x) for x in entree.split()]

    if len(donnees) < 2:
        print(" Entrez au moins 2 valeurs !")
        exit()

except ValueError:
    print(" Certaines valeurs ne sont pas des nombres !")
    exit()

afficher_rapport(donnees)
afficher_histogramme(donnees)
```
