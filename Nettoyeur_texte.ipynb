```python
#  Nettoyeur de Texte

import string   # module Python pour la ponctuation

# ================================
#  DÉFINITION DES FONCTIONS
# ================================

def supprimer_ponctuation(texte):
    resultat = ""
    for caractere in texte:
        if caractere not in string.punctuation:   # si ce n'est PAS une ponctuation
            resultat += caractere                 # on garde le caractère
    return resultat


def mettre_minuscules(texte):
    return texte.lower()


def compter_mots(texte):
    mots = texte.split()          # .split() découpe le texte en liste de mots
    return len(mots), mots        # retourne le nombre ET la liste


def nettoyer_texte(texte):
    texte_min   = mettre_minuscules(texte)        # étape 1
    texte_propre = supprimer_ponctuation(texte_min)  # étape 2
    nombre, mots = compter_mots(texte_propre)     # étape 3
    return texte_propre, nombre, mots


# ================================
# PROGRAMME PRINCIPAL
# ================================

print("=" * 40)
print("NETTOYEUR DE TEXTE")
print("=" * 40)

texte_original = input("Entrez votre texte : ")

# Nettoyage
texte_propre, nombre_mots, liste_mots = nettoyer_texte(texte_original)

# ================================
#  AFFICHAGE DES RÉSULTATS
# ================================

print("\n" + "=" * 40)
print(" RÉSULTATS")
print("=" * 40)
print(f" Texte original : {texte_original}")
print(f" Texte nettoyé  : {texte_propre}")
print(f" Nombre de mots : {nombre_mots}")
print(f" Liste des mots : {liste_mots}")
```
