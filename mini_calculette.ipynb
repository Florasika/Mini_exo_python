```python
#  Mini Calculatrice Fonctionnelle

# ================================
#  DÉFINITION DES FONCTIONS
# ================================

def additionner(a, b):
    return a + b

def soustraire(a, b):
    return a - b

def multiplier(a, b):
    return a * b

def diviser(a, b):
    if b == 0:                          #  Gestion erreur division par zéro
        return " Erreur : division par zéro impossible !"
    return a / b


# ================================
#  FONCTION PRINCIPALE
# ================================

def calculatrice():
    print("=" * 35)
    print("        MINI CALCULATRICE")
    print("=" * 35)
    print("  Opérations disponibles :")
    print("  + → Addition")
    print("  - → Soustraction")
    print("  * → Multiplication")
    print("  / → Division")
    print("=" * 35)

    #  GESTION DES ERREURS avec try/except
    try:
        a = float(input("\nEntrez le premier nombre  : "))
        b = float(input("Entrez le deuxième nombre : "))
    except ValueError:
        print(" Erreur : veuillez entrer un nombre valide !")
        return                          # on arrête la fonction

    operation = input("\nChoisissez une opération (+, -, *, /) : ")

    #  CONDITIONS POUR CHOISIR LA BONNE FONCTION
    if operation == "+":
        resultat = additionner(a, b)
        symbole  = "+"
    elif operation == "-":
        resultat = soustraire(a, b)
        symbole  = "-"
    elif operation == "*":
        resultat = multiplier(a, b)
        symbole  = "*"
    elif operation == "/":
        resultat = diviser(a, b)
        symbole  = "/"
    else:
        print(" Erreur : opération inconnue !")
        return

    #  AFFICHAGE DU RÉSULTAT
    print("\n" + "=" * 35)
    print(f"  {a} {symbole} {b} = {resultat}")
    print("=" * 35)


# ================================
#  LANCEMENT DU PROGRAMME
# ================================

calculatrice()
```
