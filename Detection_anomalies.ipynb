```python
#  Détection d'Anomalies — Z-score

import math

#  DONNÉES


donnees = [12, 15, 14, 10, 13, 11, 14, 200, 12, 9, 13, 15, -150, 11, 14]


print(" DÉTECTION D'ANOMALIES")

print(f" Données : {donnees}")
print(f" Nombre de valeurs : {len(donnees)}")


#  CALCUL MOYENNE & ÉCART-TYPE


moyenne = sum(donnees) / len(donnees)

variance = sum((x - moyenne) ** 2 for x in donnees) / len(donnees)

ecart_type = math.sqrt(variance)

print(f"\n Moyenne    : {moyenne:.2f}")
print(f" Écart-type : {ecart_type:.2f}")


#  CALCUL DU Z-SCORE


# Le z-score mesure à combien d'écarts-types une valeur
# est éloignée de la moyenne
#
#        valeur - moyenne
# z =   ─────────────────
#           écart-type

print(f"\n{'Valeur':<10} {'Z-score':<12} {'Statut'}")
print("-" * 40)

anomalies = []
normales  = []

for valeur in donnees:
    z = (valeur - moyenne) / ecart_type     # formule du z-score

    if abs(z) > 2:                          # seuil classique : 2 ou 3
        statut = " ANOMALIE"
        anomalies.append(valeur)
    else:
        statut = " Normal"
        normales.append(valeur)

    print(f"{valeur:<10} {z:<12.2f} {statut}")


#  RÉSUMÉ

print(f" Valeurs normales  ({len(normales)})  : {normales}")
print(f" Anomalies        ({len(anomalies)}) : {anomalies}")
```
