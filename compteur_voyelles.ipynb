```python
#  Compteur de voyelles

# On définit les voyelles
voyelles = "aeiouàâéèêëîïôùûü"

#  Input utilisateur
texte = input("Entrez un texte : ")

#  Dictionnaire pour compter chaque voyelle
compteur = {}

#  Boucle sur chaque lettre du texte
for lettre in texte.lower():        # .lower() → tout en minuscules
    if lettre in voyelles:          # si c'est une voyelle
        if lettre in compteur:      # si elle est déjà dans le dictionnaire
            compteur[lettre] += 1   # on ajoute 1
        else:
            compteur[lettre] = 1    # sinon on la crée avec 1

# 4️⃣ Affichage des résultats
print(f"\nTexte analysé : '{texte}'")
print(f"Nombre total de voyelles : {sum(compteur.values())}")
print("\nDétail par voyelle :")

for voyelle, nombre in compteur.items():
    print(f"  '{voyelle}' → {nombre} fois")
```
